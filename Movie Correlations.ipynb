```python
# Import the packages we will use

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Read the data
df = pd.read_csv(r'/Users/dannydelcampo/Desktop/Portfolio Project/Movies Python/movies.csv')
```

```python
df.head()
```

```python
# Check data types of columns

print(df.dtypes)
```

```python
# Sort data to see highest grossing movies

df = df.sort_values(by=['gross'], inplace = False, ascending=False)
```

```python
# Drop duplicates

df.drop_duplicates()
```

```python
# Scatter plot: Budget vs. Gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs. Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
df.head()
```

```python
# Scatter plot: Budget vs. Gross Using Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "blue"}, line_kws={"color":"green"})
```

```python
df.corr(method='pearson') #Pearson correlation 

```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)


plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
